```python
'''
# solution1: quick sort
class Solution:
    def missingNumber(self, nums: List[int]) -> int:
        def qs(nums):
            if nums == []: return []
            if len(nums) == 1: return nums
            pivot = nums[0]
            left = []
            right = []
            for num in nums[1:]:
                if num <= pivot:
                    left.append(num)
                else:
                    right.append(num)
            return qs(left) + [pivot] + qs(right)
        nums = qs(nums)
        if nums[-1] == len(nums) - 1: return nums[-1] + 1
        for i in range(len(nums)):
            if nums[i] != i: return i
'''


# solution2: missing = total(n) - sum
class Solution:
    def missingNumber(self, nums: List[int]) -> int:
        return ((len(nums)+1)*len(nums))//2 - sum(nums)
```

```python
class Solution:
    def solve(self, board: List[List[str]]) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        edge_x = len(board) - 1
        edge_y = len(board[0]) - 1
        no_surround = [[0] * len(board[0]) for _ in range(len(board))]

        def find_i(x, y):
            if x < 0 or x > edge_x or y < 0 or y > edge_y: return
            print(x, y)
            if (board[x][y] == "X") or (no_surround[x][y] == 1): return
            print(f"no_surround[{x}][{y}]=1")
            no_surround[x][y] = 1

            find_i(x-1, y)
            find_i(x+1, y)
            find_i(x, y-1)
            find_i(x, y+1)

            return

        for x in [0, edge_x]:
            for y in range(len(board[0])):
                find_i(x, y)

        for y in [0, edge_y]:
            for x in range(len(board)):
                find_i(x, y)
        print(no_surround)
        for x in range(1, edge_x):
            for y in range(1, edge_y):
                if (board[x][y] == "O") and (no_surround[x][y] == 0):
                    board[x][y] = "X"

        return
```

```python
class Solution:
    def wordPattern(self, pattern: str, s: str) -> bool:
        s = s.split()

        p_to_s = {}
        s_to_p = {}

        if len(pattern) != len(s): return False

        for i in range(len(pattern)):
            c = pattern[i]
            d = s[i]

            if c in p_to_s:
                if p_to_s[c] != d: return False
            else: p_to_s[c] = d

            if d in s_to_p:
                if s_to_p[d] != c: return False
            else: s_to_p[d] = c

        return True
```

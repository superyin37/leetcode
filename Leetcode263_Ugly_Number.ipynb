```python
class Solution:
    def isUgly(self, n: int) -> bool:
        if n <= 0: return False
        while(n % 2 == 0 and n >= 6):
            n /= 2
        while(n % 3 == 0 and n >= 6):
            n /= 3
        while(n % 5 == 0 and n >= 6):
            n /= 5
        if n <= 6: return True
        else: return False
```

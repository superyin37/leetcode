```python
class Solution:
    def summaryRanges(self, nums: List[int]) -> List[str]:
        if len(nums) == 0: return []
        if len(nums) == 1: return [f'{nums[0]}']
        i = 0
        j = 0
        result = []
        while(True):
            while(i < len(nums) - 1 and nums[i+1] - nums[i] == 1):
                i += 1
            if i == len(nums) - 1:
                result.append(f"{nums[j]}->{nums[i]}")
                break
            if i == len(nums) - 2:
                if i==j:
                    result.append(f"{nums[i]}")
                    result.append(f"{nums[i+1]}")
                    break
                if i!=j:
                    result.append(f"{nums[j]}->{nums[i]}")
                    result.append(f"{nums[i+1]}")
                    break
                result.append(f"{nums[j]}->{nums[i]}")
            if i != j:
                result.append(f"{nums[j]}->{nums[i]}")
            else:
                result.append(f"{nums[i]}")
            i = i+1
            j = i
        return result
```

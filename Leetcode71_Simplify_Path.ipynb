```python
class Solution:
    def simplifyPath(self, path: str) -> str:
        path_split = path.split("/")
        stack = []
        for s in path_split:
            if s == ".": continue
            if s == "": continue
            if s == "..":
                if not stack: continue
                else: stack.pop()
            else: stack.append(s)

        final_path = "/" + "/".join(stack)

        return final_path
```

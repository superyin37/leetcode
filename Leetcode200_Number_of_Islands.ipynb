```python
class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:
        x_edge = len(grid) - 1
        y_edge = len(grid[0]) - 1
        cnt = 0

        def flood(x, y):
            if (x<0 or x>x_edge or y<0 or y>y_edge):
                return
            # print(x, y)
            if grid[x][y] == "0": return

            grid[x][y] = "0"
            flood(x+1, y)
            flood(x-1, y)
            flood(x, y-1)
            flood(x, y+1)

        for x in range(len(grid)):
            for y in range(len(grid[0])):
                if grid[x][y] == "1":
                    cnt += 1
                flood(x, y)

        return cnt
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        if head == None or head.next == None:return True

        # step1: find middle
        fast = head.next
        slow = head
        while(fast.next != None and fast.next.next != None):
            fast = fast.next.next
            slow = slow.next
        middle = slow.next

        # step2: reverse
        node0 = None
        node1 = middle
        node2 = middle.next
        while(node1.next != None):
            node1.next = node0
            node0 = node1
            node1 = node2
            node2 = node2.next
        node1.next = node0

        # step3: compare
        tmp1 = head
        tmp2 = node1
        while(tmp2 != None):
            if tmp1.val != tmp2.val: return False
            tmp1 = tmp1.next
            tmp2 = tmp2.next
        return True
```

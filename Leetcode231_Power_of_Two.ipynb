```python
'''
# O(log(n))
class Solution:
    def isPowerOfTwo(self, n: int) -> bool:
        while(True):
            if n == 1: return True
            if n % 2 == 1: return False
            if n == 0: return False
            n /= 2
'''


# O(1) - binary represention: n & n-1 == 0
class Solution:
    def isPowerOfTwo(self, n: int) -> bool:
        if n > 0 and n & n-1 == 0: return True
        else: return False
```

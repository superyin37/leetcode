```python
# 思想1:2的幂->2进制为10000...的数
# 思想2:4的幂->1在偶数位


class Solution:
    def isPowerOfFour(self, n: int) -> bool:
        if(n > 0 and
            n & (n-1) == 0 and
            n & 0x55555555 != 0):
            return True
        else: return False
```

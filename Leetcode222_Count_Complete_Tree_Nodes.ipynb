```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

# 思路1：BFS
# Time: O(n)
# Space: O(n)
'''
from collections import deque
class Solution:
    def countNodes(self, root: Optional[TreeNode]) -> int:
        if root == None: return 0
        cnt = 0
        q = deque()
        q.append(root)
        while(q):
            tmp = q.popleft()
            cnt += 1
            if tmp.left:
                q.append(tmp.left)
                if tmp.right:
                    q.append(tmp.right)
        return cnt
'''

'''
# DFS
# Time: O(n)
# *Space: O(h)=O(log(n))
class Solution:
    def countNodes(self, root: Optional[TreeNode]) -> int:
        def DFS(root):
            if root == None:return 0
            return 1 + DFS(root.left) + DFS(root.right)
        return DFS(root)
'''


# 思路：对于任意节点，计算左右子树深度。1.深度相等则是满树，节点数2^h-1。2.深度不等则递归到左右子树
# 复杂度：对单个节点计算左右子树高度O(log n)，最大递归层数O(log n)
# 关键点：左右子树必然有一个是满的，因此总复杂度为O((log n)^2)
# Space: O(log n)

class Solution:
    def countNodes(self, root: Optional[TreeNode]) -> int:
        def countlefth(root):
            if root == None: return 0
            else: return 1 + countlefth(root.left)
        def countrighth(root):
            if root == None: return 0
            else: return 1 + countrighth(root.right)
        def countnode(root):
            if root == None: return 0
            lefth = countlefth(root)
            righth = countrighth(root)
            if lefth == righth:
                return 2**lefth - 1
            else:
                return 1 + countnode(root.left) + countnode(root.right)
        return countnode(root)
```

```python
# sort
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        result = {}

        for s in strs:
            s_sort = "".join(sorted(s))
            if s_sort not in result:
                result[s_sort] = [s]
            else:
                result[s_sort].append(s)

        return list(result.values())

# # count alphabats
# class Solution:
#     def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
#         result = {}
#         for s in strs:
#             cnt = [0] * 26
#             for ch in s:
#                 cnt[ord(ch) - ord("a")] += 1
#             key = tuple(cnt)
#             if key not in result:
#                 result[key] = [s]
#             else:
#                 result[key].append(s)
#         return list(result.values())
```
